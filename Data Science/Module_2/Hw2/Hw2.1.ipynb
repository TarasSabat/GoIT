Прочитайте дані за допомогою методу read_html з таблиці "Коефіцієнт народжуваності в регіонах України (1950—2019)" посилання (https://uk.wikipedia.org/wiki/Населення_України)



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Директива для відображення графіків в Jupyter ноутбуці
%matplotlib inline
```

```python
source_url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8"
df = pd.read_html(source_url, match="Коефіцієнт народжуваності в регіонах України", thousands=".", decimal=",")[0]
```

```python
# Виведення перших рядків першої таблиці
df.head()
```

```python
# Визначте кількість рядків та стовпців у датафреймі (атрибут shape)
rows, cols = df.shape 
print(f"Кількість рядків: {rows}")
print(f"Кількість стовпців: {cols}")
```

```python
# Замініть у таблиці значення "—" на значення NaN
df = df.replace({"—": np.nan}) 

df.head(5)
```

```python
# Визначте типи всіх стовпців за допомогою dataframe.dtypes
df.dtypes
```

```python
# Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"
nan_columns = df.columns[df.isnull().any()].tolist()
df[nan_columns] = df[nan_columns].astype(float)

df.dtypes
```

```python
# Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
df.isnull().sum() / len(df)
```

```python
# Видаліть з таблиці дані по всій країні, останній рядок таблиці
df = df.drop([27], axis=0)
df.tail(5)
```

```python
# Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)
df = df.fillna(df.mean(numeric_only=True), axis=0) 
df.head()
```

```python
# Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні
avg = df['2019'].mean()
df.loc[df['2019'] > avg]   
```

```python
# У якому регіоні була найвища народжуваність у 2014 році?

import pandas as pd
from urllib.parse import quote
import numpy as np

# Кодування URL
url = "https://uk.wikipedia.org/wiki/Населення_України"
encoded_url = quote(url, safe=":/")

# Завантаження таблиць з веб-сторінки
tables = pd.read_html(encoded_url)

# Пошук таблиці, яка містить заголовок 'Регіон'
region_with_highest_birth_rate_2014 = df[df['2014'] == df['2014'].max()]['Регіон']
region_with_highest_birth_rate_2014
```

```python
# Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році
sns.set_style('darkgrid')

plt.figure(figsize=(7, 5))
plt.bar(df['Регіон'], df['2019'], color='skyblue')
plt.xlabel('Регіон')
plt.ylabel('Народжуваність у 2019 році')
plt.title('Народжуваність по регіонах у 2019 році', fontsize=15)
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

Візуалізація

```python
# Лінійний графік
plt.figure(figsize=(20, 7))
plt.subplot(2, 3, 2)
plt.plot(df['Регіон'], df['2014'], marker='o', color='green')
plt.xlabel('Регіон')
plt.ylabel('Народжуваність у 2014 році')
plt.title('Лінійний графік народжуваності по регіонах')
plt.xticks(rotation=90)
plt.tight_layout()  # Покращення розташування елементів
plt.show()
```

```python
# Скатер-графік
plt.figure(figsize=(20, 7))
plt.subplot(2, 3, 4)
sns.scatterplot(x='Регіон', y='2014', data=df, color='purple', s=100, alpha=0.7)
plt.xlabel('Регіон')
plt.ylabel('Народжуваність у 2014 році')
plt.title('Скатер-графік народжуваності по регіонах')
plt.xticks(rotation=90)
plt.tight_layout()  # Покращення розташування елементів
plt.show()
```

```python
# Графік розподілу (Distribution Plot) для різних років
years = ['1950', '1960', '1970', '1990', '2000', '2012', '2014', '2019']
plt.figure(figsize=(20, 7))
plt.subplot(2, 3, 5)
for year in years:
        sns.kdeplot(data=df, x=year, label=year, fill=True)
plt.xlabel('Народжуваність')
plt.ylabel('Щільність')
plt.title('Графік розподілу народжуваності по роках')
plt.legend(title='Рік')
plt.tight_layout()  # Покращення розташування елементів
plt.show()
```
