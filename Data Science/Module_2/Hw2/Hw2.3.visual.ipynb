```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Директива для відображення графіків в Jupyter ноутбуці
%matplotlib inline

# Завантаження даних з CSV файлу
book_rating = pd.read_csv("bestsellers with categories.csv") 

# Перейменування колонок для зручності
book_rating.columns = ['name', 'author', 'user_rating', 'reviews', 'price', 'year', 'genre']

# Визначення статистичних показників для колонки 'price'
max_price = book_rating['price'].max()
min_price = book_rating['price'].min()
mean_price = book_rating['price'].mean()
median_price = book_rating['price'].median()
```

```python
# Лінійний графік середніх цін по роках
plt.figure(figsize=(20, 7))
plt.subplot(2, 3, 1)
average_price_per_year = book_rating.groupby('year')['price'].mean().reset_index()
plt.plot(average_price_per_year['year'], average_price_per_year['price'], marker='o', linestyle='-', color='b')
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.title('Average Book Price Per Year')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Стовпчиковий графік кількості книг по жанрам
plt.figure(figsize=(20, 7))
plt.subplot(2, 3, 2)
genre_counts = book_rating['genre'].value_counts().reset_index()
genre_counts.columns = ['genre', 'count']
sns.barplot(data=genre_counts, x='genre', y='count', hue='genre', dodge=False, palette='viridis', legend=False)
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Number of Books by Genre')

plt.tight_layout()
plt.show()
```

```python
# Boxplot цін по жанрам
plt.figure(figsize=(20, 7))
plt.subplot(2, 3, 3)
sns.boxplot(data=book_rating, x='genre', y='price', hue='genre', dodge=False, palette='pastel', legend=False)
plt.xlabel('Genre')
plt.ylabel('Price')
plt.title('Price Distribution by Genre')

plt.tight_layout()
plt.show()
```

```python
# Heatmap кореляції між числовими змінними
plt.figure(figsize=(20, 7))
plt.subplot(2, 3, 4)
corr_matrix = book_rating[['user_rating', 'reviews', 'price']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')

plt.tight_layout()
plt.show()
```

```python
# Розподіл рейтингу користувачів
plt.figure(figsize=(20, 7))
plt.subplot(2, 3, 5)
sns.histplot(book_rating['user_rating'], bins=10, kde=True, color='green')
plt.xlabel('User Rating')
plt.ylabel('Frequency')
plt.title('Distribution of User Ratings')

plt.tight_layout()
plt.show()
```
