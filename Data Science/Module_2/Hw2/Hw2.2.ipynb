Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"

```python
python_data.groupby('Посада').head()
```

Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"

```python
# Групування за стовпчиком "Посада" і агрегація мінімального та максимального значення у стовпчику "Зарплата.в.місяць"
new_developers = developers
new_developers = developers.groupby(['Посада'])['Зарплата на місяць'].agg(['min', 'max']).reset_index()
new_developers
```

Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"

```python
def fill_avg_salary(row):
    min_salary = row['min']
    max_salary = row['max']
    return (min_salary + max_salary) / 2
    
new_developers['avg'] = new_developers.apply(fill_avg_salary, axis=1)
new_developers
```

Створіть описову статистику за допомогою методу describe для нового стовпчика.

```python
# Створення описової статистики для нового стовпчика "avg"
new_developers['avg'].describe()
```

Візуалізація

```python
# Лінійний графік
plt.figure(figsize=(20, 10))
plt.subplot(2, 3, 1)
plt.plot(new_developers['Посада'], new_developers['min'], marker='o', linestyle='-', color='b', label='Min Salary')
plt.plot(new_developers['Посада'], new_developers['max'], marker='s', linestyle='--', color='r', label='Max Salary')
plt.xlabel('Посада')
plt.ylabel('Зарплата на місяць')
plt.title('Лінійний графік мінімальних та максимальних зарплат')
plt.xticks(rotation=90)
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
#Стовпчиковий графік
plt.figure(figsize=(20, 10))
plt.subplot(2, 3, 2)
new_developers.plot(kind='bar', x='Посада', y=['min', 'max'], ax=plt.gca(), color=['skyblue', 'salmon'])
plt.xlabel('Посада')
plt.ylabel('Зарплата на місяць')
plt.title('Стовпчиковий графік мінімальних та максимальних зарплат')
plt.xticks(rotation=90)
plt.legend(['Min Salary', 'Max Salary'])
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```

```python
# Графік з областю (Area plot)
plt.figure(figsize=(20, 10))
plt.subplot(2, 3, 3)
plt.fill_between(new_developers['Посада'], new_developers['min'], color="skyblue", alpha=0.4, label='Min Salary')
plt.fill_between(new_developers['Посада'], new_developers['max'], color="salmon", alpha=0.6, label='Max Salary')
plt.xlabel('Посада')
plt.ylabel('Зарплата на місяць')
plt.title('Графік з областю мінімальних та максимальних зарплат')
plt.xticks(rotation=90)
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Boxplot
plt.figure(figsize=(20, 10))
plt.subplot(2, 3, 4)
melted_data = developers[['Посада', 'Зарплата на місяць']]
sns.boxplot(data=developers, x='Посада', y='Зарплата на місяць')
plt.xlabel('Посада')
plt.ylabel('Зарплата на місяць')
plt.title('Boxplot зарплат по посадам')
plt.xticks(rotation=90)
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.3)

plt.tight_layout()
plt.show()
```
